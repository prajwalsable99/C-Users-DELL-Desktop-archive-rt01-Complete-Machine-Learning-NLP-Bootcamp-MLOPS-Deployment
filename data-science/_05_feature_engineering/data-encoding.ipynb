# label Encdoing

```python
from sklearn import preprocessing 
df=pd.read_csv('Titanic-Dataset.csv')
# label_encoder object knows  
# how to understand word labels. 
label_encoder = preprocessing.LabelEncoder() 
  
# Encode labels in column 'species'. 
df['Sex_cat']= label_encoder.fit_transform(df['Sex']) 
  
df['Sex_cat'].unique()
```

```python
df =pd.DataFrame( {'size' :["s","s","m","l","l","m","xl"] })
```

# target guided ordinal encoding

```python
df =pd.DataFrame( {'size' :["s","s","m","l","l","m","xl"] ,
                  "price": [ 200, 220,100,1,1,120,-100 ]
                  })
```

```python
df
```

```python
mean_prices= df.groupby('size')['price'].mean().to_dict()
```

```python
mean_prices
```

```python
df['size_mapped']=df['size'].map(mean_prices)
```

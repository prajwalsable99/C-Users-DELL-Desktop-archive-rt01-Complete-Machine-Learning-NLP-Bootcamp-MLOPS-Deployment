```python
import numpy as np
from sklearn.datasets import make_classification
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter
 
```

```python
# Create an imbalanced dataset
X, y = make_classification(n_classes=2, class_sep=1, weights=[0.1, 0.9],
                            n_redundant=0, flip_y=0,
                           n_features=2, n_clusters_per_class=1,
                           n_samples=1000, random_state=42)
 
print("Original class distribution:", Counter(y))
```

```python
import pandas as pd

X=pd.DataFrame( X,columns=['a','b'])
y=pd.DataFrame( y,columns=['class'])
```

```python
df= pd.concat([X,y] ,axis=1)
```

```python
df.head()
```

```python
df['class'].value_counts()
```

```python
from sklearn.utils import resample
```

```python
df_major=df[df['class']==1]
df_major.shape
```

```python
df_minor=df[df['class']==0]
df_minor.shape
```

```python
df_minor_upsampled= resample(df_minor,replace=True,n_samples=len(df_major))
```

```python
df_minor_upsampled.shape
```

```python
upsampled_df= pd.concat([df_major,df_minor_upsampled] )
```

```python
upsampled_df['class'].value_counts()
```

```python
df_major_downsampled= resample(df_major,replace=True,n_samples=len(df_minor))
```

```python
downsampled_df= pd.concat([df_major_downsampled,df_minor] )
```

```python
downsampled_df['class'].value_counts()
```

```python
df.hist()
```

```python
upsampled_df.hist()
```

```python
downsampled_df.hist()
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(df[['a','b']], df['class'])
 
# Display class distribution after SMOTE
print("Class distribution after SMOTE:", Counter(y_train_resampled))
```

```python
plt.scatter(X_train_resampled['a'],X_train_resampled['b'], c= y_train_resampled)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
from sklearn.datasets import make_blobs
```

```python
x,y=make_blobs(n_features=2,n_samples=1000,centers=3,cluster_std=1)
```

```python
plt.scatter(x[ :, 0],x[:,1])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x_train,x_test,_,_=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
wcss=[]
sill_coeffs=[]
ks= [i for i in range(2,11)]
for k in ks:
    model=KMeans(n_clusters=k,init='k-means++')
    model.fit(x_train)
    sc=silhouette_score(x_train,model.labels_)
    sill_coeffs.append(sc)
    
    wcss.append(model.inertia_)
   
```

```python
plt.plot(ks,wcss)
plt.show()
```

```python
plt.plot(ks,sill_coeffs)
plt.show()
```

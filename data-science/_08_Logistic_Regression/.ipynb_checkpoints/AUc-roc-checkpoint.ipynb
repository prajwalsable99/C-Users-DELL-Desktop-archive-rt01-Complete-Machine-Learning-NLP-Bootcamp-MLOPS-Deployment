```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_classes=2,n_features=3,n_samples=1000,n_informative=2,n_redundant=1)
```

```python
df=pd.DataFrame(x,columns=['feature1','feature2','feature3'])
```

```python
df['class']=y
```

```python
df.head()
```

```python
df.hist()
plt.show()
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python

accuracy=model.score(x_test,y_test)
accuracy
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_test,y_pred))
```

```python
#dummy model
dummy_res= [0 for i in range(len(y_test))]
```

```python
#prediction abas on probability

model_prob= model.predict_proba(x_test)
```

```python
# postive side only
model_prob[: ,1]
```

```python
from sklearn.metrics import roc_auc_score,roc_curve
```

```python
dummy_model_score=roc_auc_score(y_test,dummy_res)
ori_model_score=roc_auc_score(y_test,model_prob[:,1])

dummy_model_score,ori_model_score
```

```python
dummy_fpr,dummy_tpr,_=roc_curve(y_test,dummy_res)
ori_fpr,ori_tpr,thresholds=roc_curve(y_test,model_prob[:,1])
```

```python
plt.plot(ori_fpr,ori_tpr,c='g',label='LR model')
plt.plot(dummy_fpr,dummy_fpr,c='r',linestyle='--',label='dummy')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.legend()
plt.show()
```

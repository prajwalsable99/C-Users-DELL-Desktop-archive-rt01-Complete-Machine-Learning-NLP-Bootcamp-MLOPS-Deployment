# feature engineering

```python
df['Date']= df['Date_of_Journey'].str.split('/').str[0]
df['Month']= df['Date_of_Journey'].str.split('/').str[1]
df['Year']= df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date']= df['Date'].astype(int)
df['Month']= df['Month'].astype(int)
df['Year']= df['Year'].astype(int)
```

```python
df.drop(['Date_of_Journey'],axis=1,inplace=True)
```

```python
df.info()
```

```python

df['arrival_hour']=df['Arrival_Time'].str.split(':').str[0]
df['arrival_min']=df['Arrival_Time'].str.split(':').str[1]

```

```python
def  takefirst2char(x):
    return x[:2]
```

```python

df['arrival_min']=df['arrival_min'].apply(takefirst2char)
```

```python
df['arrival_hour']=df['arrival_hour'].astype(int)
df['arrival_min']=df['arrival_min'].astype(int)
```

```python
df.drop(['Arrival_Time'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df['dept_hour']=df['Dep_Time'].str.split(':').str[0]
df['dept_min']=df['Dep_Time'].str.split(':').str[1]

df['dept_hour']=df['dept_hour'].astype(int)
df['dept_min']=df['dept_min'].astype(int)
```

```python
df.drop(['Dep_Time'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops']=df['Total_Stops'].map( {
    
    'non-stop' :0, '2 stops':2, '1 stop':1, '3 stops':3, '4 stops':4
} )
```

```python
df.info()
```

```python
df.drop(['Route'],axis=1,inplace=True)
```

```python
li =list(df['Duration'].unique())
```

```python
def get_duration(dur):
    dur=str(dur)
    ans=0
    dur=dur.strip(' ').split(" ")
    if(len(dur)==1):
        ans= 60* int(dur[0][:-1])
    else:
        h,m= dur
        h_=60* int(h[:-1])
        m_=int(m[:-1])
        ans= h_+m_
    return ans
```

```python
df['Duration']=df['Duration'].apply(get_duration)
```

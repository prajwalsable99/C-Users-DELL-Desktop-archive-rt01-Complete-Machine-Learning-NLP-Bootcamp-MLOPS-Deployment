```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('tips.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
x=df.drop(['total_bill'],axis=1)
y=df['total_bill']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python

from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
le1=LabelEncoder()
le2=LabelEncoder()
le3=LabelEncoder()
```

```python
x_train['sex']= le1.fit_transform(x_train['sex'])
x_train['smoker']=le2.fit_transform(x_train['smoker'])
x_train['time']= le3.fit_transform(x_train['time'])
```

```python
x_train.head()
```

```python
x_test['sex']= le1.transform(x_test['sex'])
x_test['smoker']=le2.transform(x_test['smoker'])
x_test['time']= le3.transform(x_test['time'])
```

```python
x_test.head()
```

```python
x_train['day'].unique(),x_test['day'].unique()
```

```python
x
```

```python
encoder = OneHotEncoder(sparse_output=False)

# Apply one-hot encoding to the categorical columns
one_hot_encoded = encoder.fit_transform(x_train[['day']])

#Create a DataFrame with the one-hot encoded columns
#We use get_feature_names_out() to get the column names for the encoded data
one_hot_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(),index=x_train.index)
```

```python
x_train =pd.concat([x_train,one_hot_df],axis=1)
```

```python
x_train.drop(['day'],axis=1,inplace=True)
x_train.head()
```

```python
one_hot_encoded_x_test= encoder.transform(x_test[['day']])

#Create a DataFrame with the one-hot encoded columns
#We use get_feature_names_out() to get the column names for the encoded data
one_hot_xtest = pd.DataFrame(one_hot_encoded_x_test, columns=encoder.get_feature_names_out(),index=x_test.index)
```

```python

x_test =pd.concat([x_test,one_hot_xtest],axis=1)
x_test.drop(['day'],axis=1,inplace=True)
x_test.head()
```

```python
x_train.shape,x_test.shape
```

```python

from sklearn.svm import SVR
svr=SVR()
```

```python
svr.fit(x_train,y_train)
```

```python
y_pred=svr.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
```

```python
r2_score(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
# hyper parameter tuning 
```

```python
from sklearn.model_selection import GridSearchCV

params ={
 'C':[0.1,1,10,100,1000],
   'gamma':[1,0.1,0.01,0.001,0.0001],
    'kernel':['rbf','linear']
}  
```

```python
model=SVR()


grid= GridSearchCV( estimator=model, param_grid=params,cv=5,verbose=3)
```

```python
grid.fit(x_train,y_train)
```

```python
y_pred=grid.predict(x_test)
```

```python
grid.best_estimator_,grid.best_params_
```

```python
mean_squared_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```
